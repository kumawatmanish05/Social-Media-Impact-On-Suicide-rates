# Importing Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

df=pd.read_csv("/kaggle/input/impact-of-social-media-on-suicide-rates/social-media-impact-on-suicide-rates.csv")
print(df.head(15))
```

```python
print(df.info())
```

# Checking Null Values

```python
print(df.isnull().sum())
```

# EDA

**Line Plot Between Suicide Rate % change since 2010 and Per Year**

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y='Suicide Rate % change since 2010', hue='sex', marker='o')
plt.title('Suicide Rate % Change Since 2010 Over Years by Sex')
plt.show()
```

**Line Plot B/W Twitter User Count % Change Since 2010 Over Years by Sex**

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y='Twitter user count % change since 2010', hue='sex', marker='o')
plt.title('Twitter User Count % Change Since 2010 Over Years by Sex')
plt.show()
```

**Line Plot B/W Facebook User Count % Change Since 2010 Over Years by Sex**

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y='Facebook user count % change since 2010', hue='sex', marker='o')
plt.title('Facebook User Count % Change Since 2010 Over Years by Sex')
plt.show()
```

**Hist Plot B/w Suicide Rate % Change vs. Twitter User Count % Change**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Twitter user count % change since 2010', y='Suicide Rate % change since 2010', hue='sex')
plt.title('Suicide Rate % Change vs. Twitter User Count % Change')
plt.show()
```

**Scatter Plot B/w Suicide Rate % Change vs. Twitter User Count % Change**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Twitter user count % change since 2010', y='Suicide Rate % change since 2010', hue='sex')
plt.title('Suicide Rate % Change vs. Twitter User Count % Change')
plt.show()
```

**Scatter Plot B/w Suicide Rate % Change vs. Facebook User Count % Change**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Facebook user count % change since 2010', y='Suicide Rate % change since 2010', hue='sex')
plt.title('Suicide Rate % Change vs. Facebook User Count % Change')
plt.show()
```
